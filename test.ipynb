```python
from astropy.io import fits
from astropy.stats import mad_std
from wotan import flatten
import numpy as np
import matplotlib.pyplot as plt

data = fits.getdata('hlsp_tess-spoc_tess_phot_0000000001167538-s0031_tess_v1_lc.fits')

q = data['QUALITY'] == 0
t = data['TIME'][q]
f = data['PDCSAP_FLUX'][q]
flatten_lc, trend_lc = flatten(t, f, window_length=1, method='biweight', return_trend=True)
mad = mad_std(flatten_lc)
```

```python
fig, ax = plt.subplots(figsize=[13,2])
ax.plot(t, f, '.')
ax.plot(t, trend_lc, '-')
```

```python
select = 1*(np.abs(flatten_lc - 1) > 3*mad)
candidates = select > 0


aux = select[1:-1]+select[2:]+select[:-2]
findings  = aux == 3

with np.printoptions(threshold=np.inf):
    print(repr(aux).count("1, 2, 3"))

fig, ax = plt.subplots(figsize=[13,2])
ax.plot(t, flatten_lc, '.')
ax.plot(t[candidates], flatten_lc[candidates], '.')
ax.plot(t[1:-1][findings], flatten_lc[1:-1][findings], '.')
```
